```python
from tensorflow.keras.models import load_model

model = load_model("nasa_lstm_rul_model.keras")

print("Model loaded successfully!")
```

```python
import numpy as np

# Create dummy input (shape must match training input)
dummy_input = np.random.rand(1, 30, 17)  # 17 features from NASA model

prediction = model.predict(dummy_input)

print("Prediction output shape:", prediction.shape)
print("Prediction value:", prediction)
```
